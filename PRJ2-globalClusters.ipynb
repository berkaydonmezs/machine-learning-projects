```python
import numpy as np
import xarray as xr
import pandas as pd
import glob
```

```python
# open dataset
historic_path = glob.glob(r'Dataset/historic_berkeley/*')
ds = xr.open_mfdataset(historic_path)
```

```python
ds['tas'].groupby('time.month').mean()
```

```python
ds['tas'][-1000].plot()
```

```python
ds['tas'].rolling(time=12, center=True).mean()[-1000].plot()
```

```python
january_months = ds['tas'].groupby('time.month').groups[1]
january_ds = ds.isel(time=january_months)
```

```python
roll = 20
january_ds['tas'].rolling(time=roll, center=True).mean()[10].plot()
```
